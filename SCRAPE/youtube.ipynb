```python
comments = []
```

```python
def requestCom(videoId):
  request2 = youtube.commentThreads().list(
    part='snippet',
    videoId=videoId,
    textFormat='plainText',
    maxResults=200
)
  dataComment = request2.execute()
  for i in dataComment['items']:
    comments.append(i['snippet']["topLevelComment"]["snippet"]["textDisplay"])
```

```python
for i in vId:
  requestCom(i)
```

```python
df = pd.DataFrame(comments, columns=["teks"])
```
